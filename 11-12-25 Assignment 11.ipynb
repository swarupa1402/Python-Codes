<h4 style="color:blue">Question #10 – Process Intraday Data</h4> Download 5-minute data using:<br> <code>df_intra = yf.download("^NSEI", interval="5m")</code><br><br> Perform the steps:<br> 1. Rename columns to OHLCV order<br> 2. Drop the <code>"volume"</code> column<br> 3. Convert index to datetime:<br> &nbsp;&nbsp;<code>df.index = pd.to_datetime(df.index)</code><br> 4. Remove timezone:<br> &nbsp;&nbsp;<code>df.index = df.index.tz_localize(None)</code><br> 5. Format index to: <br> &nbsp;&nbsp;<code>"%Y-%m-%d %I:%M:%S"</code><br><br> Print the first 5 rows.

```python
df_intra = yf.download("^NSEI", interval="5m")
print(df_intra.head())
```

```python
#Rename columns to OHLCV order
df_intra.columns =['close', 'high', 'low', 'open', 'volume'] 

df_intra = df_intra[["open", "high", "low", "close", "volume"]]

print(df_intra.head())
```

```python
# Drop the "volume" column
df_intra = df_intra.drop(columns= ["volume"])
print(df_intra.head())
```

```python
#Convert index to datetime:
df_intra.index = pd.to_datetime(df_intra.index)
print(df_intra.head())
```

```python
#Remove timezone:
df_intra.index = df_intra.index.tz_localize(None)
print(df_intra.head())
```

```python
# Format index to:
df_intra.index = df_intra.index.strftime("%Y-%m-%d %I:%M:%S")

print(df_intra.head())
```
