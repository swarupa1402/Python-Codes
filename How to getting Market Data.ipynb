```python
import pandas as pd
import yfinance as yf
#Download the Apple price data from Jan to Dec 2019
# Set the ticker as "AAPL"
price_data_apple = yf.download ("AAPL", start = "2019-01-01", end= "2019-12-31")
#Display the last 5 rows
price_data_apple.tail()
```

```python
#for data visualisation
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
#plot the close price
plt.figure(figsize = (15,7))
price_data_apple['Close'].plot()

#set the title and axes label
plt.title('Apple Price Data', fontsize=14)
plt.xlabel('Year-Month', fontsize =12)
plt.ylabel('Price', fontsize =12)

#Show the plot
plt.show()
```

```python
import pandas as pd
import yfinance as yf

# --- Nifty 50 stock tickers on Yahoo Finance ---
nifty50_tickers = [
    "ADANIENT.NS", "ADANIPORTS.NS", "APOLLOHOSP.NS", "ASIANPAINT.NS", "AXISBANK.NS",
    "BAJAJ-AUTO.NS", "BAJFINANCE.NS", "BAJAJFINSV.NS", "BPCL.NS", "BHARTIARTL.NS",
    "BRITANNIA.NS", "CIPLA.NS", "COALINDIA.NS", "DIVISLAB.NS", "DRREDDY.NS",
    "EICHERMOT.NS", "GRASIM.NS", "HCLTECH.NS", "HDFCBANK.NS", "HDFCLIFE.NS",
    "HEROMOTOCO.NS", "HINDALCO.NS", "HINDUNILVR.NS", "ICICIBANK.NS", "ITC.NS",
    "INFY.NS", "JSWSTEEL.NS", "KOTAKBANK.NS", "LT.NS", "M&M.NS",
    "MARUTI.NS", "NTPC.NS", "ONGC.NS", "POWERGRID.NS", "RELIANCE.NS",
    "SBILIFE.NS", "SBIN.NS", "SUNPHARMA.NS", "TATACONSUM.NS", "TATAMOTORS.NS",
    "TATASTEEL.NS", "TECHM.NS", "TITAN.NS", "ULTRACEMCO.NS", "UPL.NS",
    "WIPRO.NS", "NESTLEIND.NS", "LTIM.NS", "HDFCLIFE.NS", "BAJAJFINSV.NS"
]

# --- Download live (latest) data ---
# period="1d" gives today's data; interval="1m" gives minute-level live updates
data = yf.download(tickers=nifty50_tickers, period="1d", interval="1m", group_by='ticker', threads=True)

# --- Extract latest close price for each stock ---
latest_prices = {}
for ticker in nifty50_tickers:
    try:
        latest_prices[ticker] = data[ticker]['Close'].dropna().iloc[-1]
    except Exception:
        latest_prices[ticker] = None

# --- Convert to DataFrame ---
df_live = pd.DataFrame(list(latest_prices.items()), columns=['Ticker', 'Latest Price'])

# Display
print(df_live)
```

```python
import pandas as pd
import yfinance as yf

# --- Nifty 50 tickers on Yahoo Finance ---
nifty50_tickers = [
    "ADANIENT.NS", "ADANIPORTS.NS", "APOLLOHOSP.NS", "ASIANPAINT.NS", "AXISBANK.NS",
    "BAJAJ-AUTO.NS", "BAJFINANCE.NS", "BAJAJFINSV.NS", "BPCL.NS", "BHARTIARTL.NS",
    "BRITANNIA.NS", "CIPLA.NS", "COALINDIA.NS", "DIVISLAB.NS", "DRREDDY.NS",
    "EICHERMOT.NS", "GRASIM.NS", "HCLTECH.NS", "HDFCBANK.NS", "HDFCLIFE.NS",
    "HEROMOTOCO.NS", "HINDALCO.NS", "HINDUNILVR.NS", "ICICIBANK.NS", "ITC.NS",
    "INFY.NS", "JSWSTEEL.NS", "KOTAKBANK.NS", "LT.NS", "M&M.NS",
    "MARUTI.NS", "NTPC.NS", "ONGC.NS", "POWERGRID.NS", "RELIANCE.NS",
    "SBILIFE.NS", "SBIN.NS", "SUNPHARMA.NS", "TATACONSUM.NS", "TATAMOTORS.NS",
    "TATASTEEL.NS", "TECHM.NS", "TITAN.NS", "ULTRACEMCO.NS", "UPL.NS",
    "WIPRO.NS", "NESTLEIND.NS", "LTIM.NS", "INDUSINDBK.NS", "COALINDIA.NS"
]

# --- Download October 2025 data ---
# Use start and end dates for the full month
data = yf.download(
    tickers=nifty50_tickers,
    start="2025-10-01",
    end="2025-10-31",
    group_by='ticker',
    threads=True
)

# --- Optional: show last few rows of one company ---
print(data['RELIANCE.NS'].tail())

# --- Save to Excel file for reference ---
output_file =r"C:\Users\Swarupa\Downloads\Nifty50_October2025_Data.xlsx"
with pd.ExcelWriter(output_file) as writer:
    for ticker in nifty50_tickers:
        if ticker in data.columns.get_level_values(0):
            data[ticker].to_excel(writer, sheet_name=ticker.replace('.NS', ''))

print(f"\n Saved all October 2025 data to: {output_file}")
```
