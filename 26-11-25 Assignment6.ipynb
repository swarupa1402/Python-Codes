<h4 style="color:blue">Question #1 - Basic If condition</h4> 
Write a program where score = 75 and print:<br>
"Pass" if score is above 60, "Fail" otherwise<br>

<h4 style="color:blue">Question #2 - If–Elif–Else</h4> 
Given a variable marks = 82, print:

“A Grade” if marks > 90

“B Grade” if marks > 75

“C Grade” otherwise

```python
marks = 82
if marks > 90:
    print("A Grade")
elif marks > 75:
    print("B Grade")
else:
    print("C Grade")
```

<h4 style="color:blue">Question #4 - Loop Over List</h4> 
prices = [68, 72, 70, 75, 65]

For each price, print:

“buy” if price > 70

“sell” if price < 70

“hold” if price == 70

```python
prices = [68, 72, 70, 75, 65]

for i in prices:
    if i > 70:
        print("buy")
    elif i < 70:
        print("sell")
    elif i == 70:
        print("hold")
```

<h4 style="color:blue">Question #6 - For Loop, Continue</h4> 
Using a for loop from 1 to 6, print all numbers except 4 using continue

```python
for i in range (1,7):
    if i == 4:
        continue
    print(i)
```

<h4 style="color:blue">Question #7 - For Loop, Break</h4> 
Using a for loop from 1 to 10, stop the loop when the value becomes 6.

```python
for i in range (1, 10):
    if i == 6:
        break
    print(i)
```

<h4 style="color:blue">Question #9 - Zip function</h4> 
Given:<br>
stocks = ["AAPL", "MSFT", "TSLA"]<br>
prices = [193.5, 108.2, 252.1]<br>
Use zip() and print each stock with its price.

```python
stocks = ["AAPL", "MSFT", "TSLA"]
prices = [193.5, 108.2, 252.1]

z = zip(stocks , prices)
print(z)

list_z = list(z)
print(list_z)
```

<h4 style="color:blue">Question #10 - Generate Buy/Sell signal </h4> 
short_ma = [20, 22, 19, 25, 24]<br>
long_ma  = [18, 21, 20, 23, 26]

Write a loop that compares values at each position and prints:

“Buy” if short_ma[i] > long_ma[i]

“Sell” if short_ma[i] < long_ma[i]

```python
short_ma = [20, 22, 19, 25, 24]
long_ma = [18, 21, 20, 23, 26]

for i in range (0, len(short_ma)):
    if short_ma[i] > long_ma[i]:
        print("Buy")
    elif short_ma[i] < long_ma[i]:
        print("Sell")
```
