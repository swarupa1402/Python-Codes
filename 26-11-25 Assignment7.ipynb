<h4 style="color:blue">Question #1 - Candle Type Identification</h4> Using the lists below:<br> open_price = [220, 225, 223, 230, 228, 232, 235, 240]<br> close_price = [224, 220, 226, 231, 227, 236, 233, 242]<br><br> Write a function <b>candle_type(open_p, close_p)</b> that returns:<br> "Bullish" / "Bearish" / "Doji"<br> Loop using <b>zip()</b> and print the full list of candle types.

```python
open_price = [220, 225, 223, 230, 228, 232, 235, 240]
close_price = [224, 220, 226, 231, 227, 236, 233, 242]
def candle_type(open_p, close_p):
    if open_p > close_p:
        return "Bullish"
    elif open_p < close_p:
        return"Bearish"
    else:
        return"Doji"
        
candle_list = []

for x, y in zip(open_price, close_price):
    candle_list.append(candle_type(x, y))

print(candle_list)
```

<h4 style="color:blue">Question #2 - Percentage Change List</h4> Given:<br> prices = [150, 152, 149, 155, 160, 158, 162, 165]<br><br> Write a function <b>pct_change(listname)</b> that returns a list of percentage changes.<br> First value should be <b>None</b>.

```python
prices = [150, 152, 149, 155, 160, 158, 162, 165]

def pct_change(price_list):
    pct_values = []
    
    for i in range(len(price_list)):
        if i == 0:
            pct_values.append(None)   
        else:
            prev = price_list[i-1]
            curr = price_list[i]
            pct = ((curr - prev) / prev) * 100
            pct_values.append(round(pct, 2))
    
    return pct_values

pct_vals = pct_change(prices)
print(pct_vals )
```

<h4 style="color:blue">Question #3 - OHLC Range & Mid Price</h4> Given:<br> highs = [305, 310, 308, 315, 320]<br> lows = [300, 305, 304, 310, 315]<br><br> Write a function <b>range_mid(high, low)</b> that returns:<br> • Range (high - low)<br> • Mid price ((high + low) / 2)<br><br> Store all results in two lists: <b>range_list</b> and <b>mid_list</b>.

```python
highs = [305, 310, 308, 315, 320]
lows = [300, 305, 304, 310, 315]

def range_mid(high, low):
    range_list = high - low
    mid_list = (high + low) / 2
    return range_list, mid_list

range_list = []
mid_list = []

for h, l in zip(highs, lows):
    r, m = range_mid(h, l)
    range_list.append(r)
    mid_list.append(m)

print("Range List:", range_list)
print("Mid List:", mid_list)
```

<h4 style="color:blue">Question #5 - Generate Buy/Sell Signals</h4> Given:<br> close_list = [105, 102, 108, 110, 107, 112]<br> sma_list = [103, 104, 105, 109, 108, 111]<br><br> Write a function <b>generate_signal(close, sma)</b> that returns:<br> "Buy" if close > sma, otherwise "Sell".<br> Loop and print all signals.

<h4 style="color:blue">Question #6 - Calculate Spreads Using map()</h4> Given:<br> bid = [100, 102, 101, 105]<br> ask = [101, 103, 102, 106]<br><br> Use <b>map() + lambda</b> to compute the spread:<br> ask - bid<br> Return the list of spreads.

```python
bid = [100, 102, 101, 105]
ask = [101, 103, 102, 106]

spreads = list(map(lambda b, a: a - b, bid, ask))

print(spreads)

```

<h4 style="color:blue">Question #8 - SL & TP Function (Default Parameters)</h4> Write a function:<br> <b>def sl_tp(price, sl=0.03, tp=0.06):</b><br> Return the SL and TP levels.<br><br>

Given:<br>
entry_prices = [1000, 1020, 980, 1100]<br><br>
Loop and print SL & TP for each price.

```python
def sl_tp(price, sl=0.03, tp=0.06):
    stop_loss = price * (1 - sl)
    take_profit = price * (1 + tp)
    return stop_loss, take_profit

entry_prices = [1000, 1020, 980, 1100]

for price in entry_prices:
    sl, tp = sl_tp(price)
    print(f"Entry: {price} -> SL: {sl:.2f}, TP: {tp:.2f}")

```

<h4 style="color:blue">Question #9 - Trend Checker Using zip()</h4> Given:<br> day1 = [140, 141, 143, 142, 145]<br> day2 = [142, 140, 144, 146, 147]<br><br> Write a function <b>trend(a, b)</b> that returns:<br> "Uptrend" if b > a, else "Downtrend".<br> Use <b>zip()</b> to print the full trend list.

```python
day1 = [140, 141, 143, 142, 145]
day2 = [142, 140, 144, 146, 147]

def trend(a, b):
    if b > a:
        return "Uptrend"
    else:
        return "Downtrend"

trend_list = []

for d1, d2 in zip(day1, day2):
    trend_list.append(trend(d1, d2))

print(trend_list)
```

<h4 style="color:blue">Question #10 - Count Buy & Sell Using filter()</h4> Given:<br> signals = ["Buy", "Sell", "Buy", "Buy", "Sell", "Sell", "Buy"]<br><br> Use <b>filter()</b> to:<br> • Extract only all "Buy"<br> • Extract only all "Sell"<br> Print how many Buy and Sell signals are present.

```python
signals = ["Buy", "Sell", "Buy", "Buy", "Sell", "Sell", "Buy"]

buy_signals = list(filter(lambda s: s == "Buy", signals))
sell_signals = list(filter(lambda s: s == "Sell", signals))

print("Buy Signals:", buy_signals)
print("Sell Signals:", sell_signals)
print("Total Buy:", len(buy_signals))
print("Total Sell:", len(sell_signals))
```
