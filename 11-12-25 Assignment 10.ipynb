<h4 style="color:blue">Question #1 – Create and Print a Simple Pandas Series</h4> Create a Series <b>S1</b> using:<br> <code>S1 = pd.Series([15, 30, 45, 60, 75])</code><br><br> Print:<br> • The Series<br> • The index<br> • The values<br>

```python
import pandas as pd

S1 = pd.Series([15, 30, 45, 60, 75])
print("Series:\n",S1)
print("Index:\n", S1.index)
print("Values:\n", S1.values)
```

<h4 style="color:blue">Question #4 – Create Fruit Price Series</h4> Create a Series <b>F1</b> using fruit names as index and prices as values:<br> <code> fruits = ['Apple','Banana','Orange','Grapes']<br> F1 = pd.Series([120, 40, 60, 150], index=fruits) </code><br><br> Print:<br> • The Series<br> • The price of <b>Orange</b><br>

<h4 style="color:blue">Question #6 – Add Series with Mismatched Index</h4> Create:<br> <code> S1 = pd.Series([4000, 4500, 5000, 5500], index=['Jan','Feb','Mar','Apr'])<br> S2 = pd.Series([4200, 4800, 5100], index=['Feb','Mar','Apr']) </code><br><br> Add them and print:<br> • The result<br> • Missing values using <code>(S1 + S2).isnull()</code><br>

```python
S1 = pd.Series([4000, 4500, 5000, 5500], index=['Jan','Feb','Mar','Apr'])
S2 = pd.Series([4200, 4800, 5100], index=['Feb','Mar','Apr']) 

result = S1 + S2

print("Result:\n", result)
print("Missing Values:\n", result.isnull())
```

<h4 style="color:blue">Question #7 – Drop Missing Values</h4> Using the result of <code>S1 + S2</code> from Question 6, print:<br> • The cleaned Series using <code>.dropna()</code><br>

```python
cleaned = result.dropna()
print(cleaned)
```

<h4 style="color:blue">Question #8 – Replace Missing Values</h4> Using the same result from Question 6, replace missing values with <b>0</b>:<br> <code>(S1 + S2).fillna(0)</code><br>

```python
filled = result.fillna(0)
print(filled)
```

<h4 style="color:blue">Question #9 – Apply NumPy Functions on a Series</h4> Create a Series <b>temp</b> representing temperatures:<br> <code>temp = pd.Series([25, 28, 30, 32, 35])</code><br><br> Apply and print:<br> • <code>np.sqrt(temp)</code><br> • <code>np.log(temp)</code><br>

```python
import numpy as np

temp = pd.Series([25, 28, 30, 32, 35])

print("Square root\n",np.sqrt(temp))
print("Log values\n",np.log(temp))
```

<h4 style="color:blue">Question #10 – Create Series from a Dictionary</h4> Given the dictionary:<br> <code> cars = {'Hyundai':900000, 'Maruti':700000, 'Honda':1100000, 'Tata':850000} </code><br><br> Create a Series <b>C</b> and print:<br> • The entire Series<br> • The price of <b>Honda</b><br> • Prices of <b>Maruti</b> and <b>Tata</b> using list indexing<br>

```python
cars = {'Hyundai':900000, 'Maruti':700000, 'Honda':1100000, 'Tata':850000} 

C = pd.Series(cars)

print("Series:\n", C)
print("Price of Honda:\n", C['Honda'])
print("Price of Maruti and Tata:\n",C [['Maruti', 'Tata']])
```
