# üß± Python Data Structures

### Data structures are different ways of storing and organizing data so that it can be used efficiently.
Python provides several built-in data structures that make it easy to handle different types of information.<br>
Each structure has its own features, strengths, and use cases - depending on whether you need to store ordered data, unique values, or key‚Äìvalue pairs.

## The 4 main data structures in Python are:

**List** ‚Äì an ordered, changeable collection of items. 

**Tuple** ‚Äì an ordered, unchangeable (immutable) collection.

**Set** ‚Äì an unordered collection of unique items.

**Dictionary** ‚Äì a collection of key‚Äìvalue pairs for quick lookups.

### Ordered: Items have a fixed position (index) and always appear in the same sequence when you access or print them.

### Mutable: You can change, add, or remove elements after the object is created.

### Allow Duplicates:The same value can appear more than once inside the collection.

# Dictionaries { }

A dictionary in Python is used to store data in pairs ‚Äî called key‚Äìvalue pairs.
Each key is unique and acts like a label that helps you quickly find its corresponding value.
Dictionaries are unordered, changeable (mutable), and written using curly braces { }

### A key in a dictionary is the unique identifier used to access a specific value

```python
new_dict = {}  # Empty Dictionary

type(new_dict)
```

```python
student = {"name": "Amit", "age": 20, "course": "Python" }

print(student["name"])     # Access value using key
```

```python
print(student)
```

```python
student["age"] = 21        # Update a value
print(student)
```

```python
student["grade"] = "A"     # Add a new key‚Äìvalue pair
```

```python
print(student)
```

<b>sorted</b> (x_dict) <br>

The dictionary will get sorted by its values.

```python
student_scores = {"Rahul": 85, "Amit": 92, "Neha": 78, "Sneha": 88}

# Sort dictionary by keys (names)
sorted_names = sorted(student_scores)

print(sorted_names)
```

If you want to sort by the values, you can do this:

```python
sorted_by_value = sorted(student_scores, key=student_scores.get)
print(sorted_by_value) #Keys sorted by Values
```

```python
# Sort by values (ascending) and print all key + value pairs

for name in sorted(student_scores, key=student_scores.get):
    print(name, ":", student_scores[name])
```

```python
# Sort by values (ascending) and print all key + value pairs
for name in sorted(student_scores, key=student_scores.get, reverse=True):
    print(name, ":", student_scores[name])
```

# Sets

A set is an unordered collection with no duplicate elements. They are useful to create lists that hold only unique values and are also mutable. The elements of a set can be anything like numbers, strings or characters.

## How to create Sets
Curly braces or the set () function can be used to create sets and the items within them are separated by commas.

```python
my_set = set() # this is an empty set
print(my_set)
type(my_set)
```

```python
# We can use the set() function to create sets
set_a = set('TEMPERATUREEEEE') #contains 11 words
print(type(set_a))
print(set_a) #only unique values will b
```

### üîπ Key idea - Sets are not ordered
**Sets only care what elements are present, not the order in which they appear.<br>
So, there‚Äôs no concept of ‚Äúfirst‚Äù or ‚Äúlast‚Äù element in a set.**<br>
**Only Unique values are printed**

```python
A = {10, 20, 30, 40}
B = {30, 10, 20}
print(A == B)   # Only contents are compared not positions
```

## Set operations

Mathematical operations like set union, set intersection, set difference can be performed

```python
# Lets create 2 sets - set_1 and set_2

set_1 = set('LMKNOP')
set_2 = set('OPQRST')

print(set_1)
print(set_2)
```

<b> x.union(y) </b> <br>
This method returns all the unique items that are present in the two sets, as a new set.

```python
set_1.union(set_2)
```

```python
#Simpler alternate way of performing Union is by using the pipe '|' operator
set_1 | set_2
```

<b> x.intersection(y) </b> <br>
This method returns the common items that are present in two sets, as a new set.

```python
set_1.intersection(set_2)
```

```python
#Simpler alternate way of performing Intersection is by using ampersand '&' operator
set_1 & set_2
```

<b> x.difference(y) </b> <br>
This method returns the items of 'set 1' which are not common to the 'set 2', as a new set. 

```python
print(set_1)
print(set_2)
```

```python
set_1.difference(set_2)
```

```python
#Simpler alternate way of performing Difference is by using minus '-' operator
set_1 - set_2
```

<b> difference_update () </b> <br>
Removes all elements from the first set that are also present in the second set.<br>
It modifies the first set - no new set is created.

```python
set_1.difference_update(set_2)
print(set_1)
print(set_2)
```

<b>x.isdisjoint(y)</b> <br>
The method checks whether two sets have no elements in common.<br>
**Returns:**<br>
True ‚Üí if the sets have no common elements<br>
False ‚Üí if any element is shared between them

```python
set_1.isdisjoint(set_2)
```

```python
set_3 = set('ABCD')
set_4 = set('CDEF')
print(set_3)
print(set_4)
```

```python
set_3.isdisjoint(set_4) #Here we have common elements
```

<b>y.issubset(x)</b> <br>
This method returns True for 'Set 2', if all the elements of 'Set 2' are present in 'Set 1'

### Easy way to remember:
**Subset ‚Üí smaller inside bigger**<br>
**Superset ‚Üí bigger contains smaller**

```python
set_5 = set('ABCDEFG')
set_6 = set('EFG')
print(set_5)
print(set_6)
```

```python
set_6.issubset(set_5)
```

**Because set_5 has extra elements (A, B, C, D) that are not present in set_6.
So, set_5 is not a subset of set_6.**

```python
set_5.issubset(set_6) 
```
